```python
import pandas as pd

# Load the datasets
txt_file = r"C:\Users\Siddhant Nijhawan\Downloads\NEST\Problem Statements and Data Sets\drop_withdrawals.txt"
csv_file = r"C:\Users\Siddhant Nijhawan\Downloads\NEST\Problem Statements and Data Sets\usecase_3_.csv"

df_txt = pd.read_csv(txt_file, delimiter="|")
df_csv = pd.read_csv(csv_file)

# Select required columns from CSV
df_csv_subset = df_csv[['NCT Number', 'Study Status']]

# Perform the join
merged_df = pd.merge(
    df_txt, 
    df_csv_subset, 
    how="left", 
    left_on="nct_id", 
    right_on="NCT Number"
)

# Display the merged DataFrame
print(merged_df)
```

```python
pivot_table = pd.pivot_table(
    merged_df,
    values='count',          
    index=['period'],       
    columns=['Study Status'],  
    aggfunc='count',          
    fill_value=0             
)

print(pivot_table)
```

```python
sorted_pivot_table = pivot_table.sort_values(by='COMPLETED', ascending=False)

print(sorted_pivot_table)
```

```python
import pandas as pd

# Assuming 'sorted_pivot_table' is already created as per the earlier step
# Export to Excel
output_file = r"C:\Users\Siddhant Nijhawan\Downloads\NEST\new_excel111.xlsx"
sorted_pivot_table.to_excel(output_file)

print(f"Pivot table has been exported to: {output_file}")
```

```python
import os
import pandas as pd

# Define the file paths
file_paths = [
    r"C:\Users\Siddhant Nijhawan\Downloads\NEST\Problem Statements and Data Sets\reported_events.txt",
    r"C:\Users\Siddhant Nijhawan\Downloads\NEST\Problem Statements and Data Sets\facilities.txt",
    r"C:\Users\Siddhant Nijhawan\Downloads\NEST\Problem Statements and Data Sets\drop_withdrawals.txt"
]

# Create a new directory for the output files
output_dir = r"C:\Users\Siddhant Nijhawan\Downloads\NEST\Problem Statements and Data Sets\ConvertedFiles"
os.makedirs(output_dir, exist_ok=True)

# Convert each file to CSV
for file_path in file_paths:
    # Read the file into a DataFrame
    df = pd.read_csv(file_path, delimiter='|')
    
    # Extract the base filename and replace extension
    base_name = os.path.basename(file_path).replace('.txt', '.csv')
    
    # Define the output path
    output_path = os.path.join(output_dir, base_name)
    
    # Save DataFrame as CSV
    df.to_csv(output_path, index=False)
    print(f"Converted {file_path} to {output_path}")
```
