```python
import pandas as pd


def classify_period(cumsum_value, period_name):
    if cumsum_value <= 71:
        return period_name  # retaining original period name
    elif 72 <= cumsum_value <= 80:
        return '72-80%'
    elif 81 <= cumsum_value <= 90:
        return '81-90%'
    else:
        return '91-100%'

df1['period_class'] = df1.apply(
    lambda row: classify_period((row['cumulative_frequency_percentage'] / df1['cumulative_frequency_percentage'].max()) * 100, row['period']),
    axis=1
)

print(df1[['period', 'cumulative_frequency_percentage', 'period_class']])
```

Merging the txt files one by one with the usecase3 csv

Starting with `faciltities.txt`

This code performs a series of data processing steps on the country column of the final_result DataFrame to ensure consistent categorization and analysis:

1. Merging Country Data:

The book1 DataFrame is merged with the dummy_data DataFrame on the NCT Number field, aggregating the country data into a single column for each unique NCT Number. Countries are concatenated using a semicolon (;) as a separator.

2. Retaining Specific Countries:

A list of important countries (retain_countries) is defined based on the analysis. For countries not in this list, their values are replaced with "countries with low cumsum value".

3. One-Hot Encoding: Converts the country column into binary columns (country_<name>) for each unique country, enabling easier analysis.

```python
import pandas as pd

book1 = df
dummy_data = df3

merged_data = book1.merge(dummy_data, left_on="NCT Number", right_on="nct_id", how="left")

result = (
    merged_data.groupby("NCT Number")["country"]
    .apply(lambda x: "; ".join(x.dropna().unique()) if x.notna().any() else None)
    .reset_index()
)

final_result = book1.merge(result, on="NCT Number", how="left")
```

```python
final_result["country"].head(15)
```

```python
# country names to retain according to the above analysis
retain_countries = [
    "United States", "Germany", "France", "Japan", "Canada", "Spain", 
    "United Kingdom", "Italy", "China", "Poland", "Russian Federation", 
    "Korea, Republic of", "Belgium", "Australia", "Netherlands", "Brazil", 
    "Turkey", "Hungary", "India", "Sweden", "Ukraine", "Denmark", 
    "Argentina", "Israel", "Taiwan"
]

def modify_countries(countries):
    if pd.isna(countries):  
        return None
    country_list = countries.split("; ")
    updated_countries = [
        country if country in retain_countries else "countries with low cumsum value" 
        for country in country_list
    ]
    return "; ".join(updated_countries)

final_result["country"] = final_result["country"].apply(modify_countries)

final_result["country"].head(10)
```

```python
def one_hot_encode_countries(dataframe, column, prefix):
    dataframe[column] = dataframe[column].fillna("")
    
    dataframe["country_list"] = dataframe[column].apply(lambda x: x.split("; ") if x else [])
    
    all_countries = set(
        country for sublist in dataframe["country_list"] for country in sublist if country
    )
    
    for country in all_countries:
        col_name = f"{prefix}_{country}"
        dataframe[col_name] = dataframe["country_list"].apply(lambda x: 1 if country in x else 0)
    
    dataframe[column] = dataframe[column].replace("", None)
    dataframe.drop(columns=["country_list"], inplace=True)

# Applying one-hot encoding to the 'country' column with the prefix 'country'
one_hot_encode_countries(final_result, column="country", prefix="country")

final_result
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sentence_transformers import SentenceTransformer
import umap
import hdbscan
import matplotlib.pyplot as plt
from collections import Counter

all_events = final_result5['adverse_event_term'].str.split(';').explode() 
all_events = all_events.str.strip().str.lower()  
unique_events = all_events.dropna().unique()  
print(f"Number of unique diseases: {len(unique_events)}")

# Encoding the terms using a Transformer
print("Encoding text with a pre-trained Transformer...")
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2') 
encoded_events = model.encode(unique_events, show_progress_bar=True)

print("Reducing dimensionality with UMAP...")
umap_reducer = umap.UMAP(n_neighbors=15, n_components=10, metric='cosine', random_state=42)
umap_embeddings = umap_reducer.fit_transform(encoded_events)

print("Clustering with HDBSCAN...")
clusterer = hdbscan.HDBSCAN(min_cluster_size=50, metric='euclidean', cluster_selection_epsilon=0.1)
cluster_labels = clusterer.fit_predict(umap_embeddings)

unique_clusters = np.unique(cluster_labels)
print(f"Number of clusters found: {len(unique_clusters)} (including noise cluster: -1)")
cluster_counts = Counter(cluster_labels)
for cluster, count in cluster_counts.items():
    print(f"Cluster {cluster}: {count} items")

print("Visualizing clusters...")
plt.scatter(umap_embeddings[:, 0], umap_embeddings[:, 1], c=cluster_labels, cmap='Spectral', s=5)
plt.colorbar()
plt.title("UMAP Projection with HDBSCAN Clusters")
plt.show()

print("Replacing text entries with cluster labels...")
event_to_cluster = {event: cluster for event, cluster in zip(unique_events, cluster_labels)}
final_result5['adverse_event_cluster'] = final_result5['adverse_event_term'].apply(
    lambda x: [event_to_cluster.get(event.strip().lower(), -1) for event in x.split(';') if event.strip()]
)
print(final_result5[['adverse_event_term', 'adverse_event_cluster']].head())
```

```python
def classify_cluster(cumulative_percentage, cluster_name):
    if cumulative_percentage is None:
        return None
    
    if isinstance(cumulative_percentage, list):
        if all(x < 70 for x in cumulative_percentage):
            return cluster_name  
        elif any(70 <= x <= 80 for x in cumulative_percentage):
            return '70-80%'
        elif any(81 <= x <= 90 for x in cumulative_percentage):
            return '81-90%'
        else:
            return '91-100%'
    elif cumulative_percentage < 70:
        return cluster_name
    elif 70 <= cumulative_percentage <= 80:
        return '70-80%'
    elif 81 <= cumulative_percentage <= 90:
        return '81-90%'
    else:
        return '91-100%'

# Applying classification to `final_result5` to create `adverse_event_cluster_type`
final_result5['adverse_event_cluster_type'] = final_result5.apply(
    lambda row: classify_cluster(row['cumulative_frequency_percentage'], row['adverse_event_cluster']),
    axis=1
)

# One-hot encoding
unique_categories = final_result5['adverse_event_cluster_type'].explode().dropna().unique()

for category in unique_categories:
    column_name = f"_adverse_event_cluster_{category}"
    final_result5[column_name] = final_result5['adverse_event_cluster_type'].apply(
        lambda x: 1 if isinstance(x, list) and category in x else 0
    )
print("Updated final_result5 with prefixed column names:")
```
