Outlier Detection and Capping with Visualization

This code uses the IQR method to detect and handle outliers in specified columns:

1. Detection: Identifies values outside 1.5 * IQR as outliers.
2. Capping: Replaces outliers with the nearest bound (lower or upper).
3. Visualization: Box plots show distributions before and after capping.

This ensures a robust dataset by minimizing the impact of extreme values.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

columns = ['Enrollment', 'subjects_at_risk', 'subjects_affected', 'duration',
           'minimum_age', 'child', 'maximum_age', 'older_adult', 'adult']

def detect_outliers(df, column):
    Q1 = df[column].quantile(0.25)  
    Q3 = df[column].quantile(0.75)  
    IQR = Q3 - Q1                   
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers, lower_bound, upper_bound

def cap_outliers(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])
    return df

for col in columns:
    outliers, lower, upper = detect_outliers(sidd, col)
    print(f"Column: {col}")
    print(f"Lower Bound: {lower}, Upper Bound: {upper}")
    print(f"Number of Outliers: {len(outliers)}\n")

    plt.figure()
    sidd[col].plot(kind='box')
    plt.title(f"Before Capping: {col}")
    plt.show()

    sidd = cap_outliers(sidd, col)

    plt.figure()
    sidd[col].plot(kind='box')
    plt.title(f"After Capping: {col}")

    plt.show()
```

cap outliers based on custom thresholds or percentiles

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

columns = ['Enrollment', 'subjects_at_risk', 'subjects_affected', 'duration',
           'minimum_age', 'child', 'maximum_age', 'older_adult', 'adult']

def cap_outliers_custom(df, column, lower_bound=None, upper_bound=None):
    if lower_bound is not None:
        df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
    if upper_bound is not None:
        df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])
    return df

for col in columns:
    if col in ['Enrollment', 'duration']:
        #caping using percentiles (1st and 99th)
        lower = np.percentile(sidd[col], 1)
        upper = np.percentile(sidd[col], 99)
        sidd = cap_outliers_custom(sidd, col, lower, upper)
        print(f"Capped {col} between {lower} and {upper}")

    elif col in ['minimum_age', 'maximum_age']:
        #using logical bounds
        if col == 'minimum_age':
            sidd = cap_outliers_custom(sidd, col, lower_bound=18)
        elif col == 'maximum_age':
            sidd = cap_outliers_custom(sidd, col, lower_bound=18, upper_bound=100)  
        print(f"Processed {col} with custom bounds")

    else:
        print(f"Retained {col} without changes")

    plt.figure()
    sidd[col].plot(kind='box')
    plt.title(f"After Processing: {col}")
    plt.show()

skewed_columns = ['Enrollment', 'subjects_at_risk', 'subjects_affected', 'duration']
for col in skewed_columns:
    sidd[col] = np.log1p(sidd[col])  #Log transform to reduce skewness
    print(f"Applied log transformation to {col}")

print(sidd.describe())
```
